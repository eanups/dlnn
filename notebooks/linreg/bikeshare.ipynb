# Building a Linear Regression Model

```python
feature_cols = ['temp']
X = bikes[feature_cols]
y = bikes.total
```

```python
from sklearn.linear_model import LinearRegression
```

```python
linreg = LinearRegression()
linreg.fit(X, y)
```

```python
print linreg.intercept_
print linreg.coef_
```

Interpreting the intercept (0):

It is the value of y when x=0.
Thus, it is the estimated number of rentals when the temperature is 0 degrees Celsius.
Note: It does not always make sense to interpret the intercept. (Why?)
Interpreting the "temp" coefficient (1):

It is the change in y divided by change in x, or the "slope".
Thus, a temperature increase of 1 degree Celsius is associated with a rental increase of 9.17 bikes.
This is not a statement of causation.
1 would be negative if an increase in temperature was associated with a decrease in rentals.

### How many bikes if an increase in bikes to 25 ?

```python
linreg.intercept_ + 25 * linreg.coef_
```

```python
linreg.predict(25)
```

Errors

```python
# sample
actual = [12, 33, 14, 22]
predicted = [14, 26, 20, 20]
```

```python
from sklearn import metrics
import numpy as np

print 'MAE: ', metrics.mean_absolute_error(y_true=actual, y_pred=predicted)
mse = metrics.mean_squared_error(actual, predicted)
print 'MSE: ', mse
print 'RMSE: ', np.sqrt(mse)
```

```python
# Function to accept a feature set and returns RMSE
from sklearn.model_selection import train_test_split


def train_test_features(features):
    X = bikes[features]
    y = bikes.total
    (X_train, X_test, y_train, y_test) = train_test_split(X, y, random_state=123)
    linreg = LinearRegression()
    linreg.fit(X_train, y_train)
    y_predicted = linreg.predict(X_test)
    return np.sqrt(metrics.mean_squared_error(y_test, y_predicted))
```

```python
# compare different sets of features
print train_test_features(['temp', 'season', 'weather', 'humidity'])
print train_test_features(['temp', 'season', 'weather'])
print train_test_features(['temp', 'season', 'humidity'])
```
